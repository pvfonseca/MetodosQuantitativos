# Regressão linear

* Base de dados ``cars`` RDatasets
* DataFrame com duas colunas: ``speed`` e ``dist``
* Os dados mostram a velocidade dos carros (mph) e as distâncias necessárias para parada (ft).

## Como ``speed`` e ``dist`` estão relacionadas?

```julia
cars = dataset("datasets","cars")
```

```julia
plot(cars[!, "Speed"], cars[!, "Dist"], seriestype=:scatter, size=(800,600), mc=:red, label=:none, ms=5, msc=:red)
title!("Stopping distance vs Speed")
xlabel!("Speed (mph)")
ylabel!("Stopping distance (ft)")
```

> Como ``speed`` e ``dist`` estão relacionadas? Qual a melhor forma de resumir esta relação?

* Podemos ajustar uma linha horizontal pelo diagrama de dispersão:

```julia
plot(cars[!, "Speed"], cars[!, "Dist"], seriestype=:scatter, size=(800,600), mc=:red, label=:none, ms=5, msc=:red)
title!("Stopping distance vs Speed")
xlabel!("Speed (mph)")
ylabel!("Stopping distance (ft)")
hline!([60], label=:none, lc=:black, lw=3)
```

> Note que esta não é uma forma apropriada para resumir a relação.
>
> Claramente, a *melhor* linha não seria horizontal mas, sim, teria uma *inclinação* positiva: 

```julia
plot(cars[!, "Speed"], cars[!, "Dist"], seriestype=:scatter, size=(800,600), mc=:red, label=:none, ms=5, msc=:red)
title!("Stopping distance vs Speed")
xlabel!("Speed (mph)")
ylabel!("Stopping distance (ft)")
plot!(cars[!, "Speed"], 5*cars[!, "Speed"], label=:none, lc=:black, lw=3)
```

> Representação um pouco melhor. No entanto, a linha parece estar em um nível elevado (intercepto muito alto)]
>
> Nosso objetivo é encontrar a relação linear entre as variáveis ``speed`` e ``dist`` que melhor se ajusta aos dados
>
> Dito de outra forma, quais os valores de $\beta_0$ e $\beta_1$ que melhor ajusta a seguinte relação linear ao diagrama de dispersão?
>
> $$z = \beta_0 + \beta_1 x$$

```julia
# Inclinação
β1 = cov(cars[!, "Speed"], cars[!, "Dist"])/var(cars[!, "Speed"])
```

```julia
# Intercepto
β0 = mean(cars[!, "Dist"]) - β1*mean(cars[!, "Speed"])
```

```julia
plot(cars[!, "Speed"], cars[!, "Dist"], seriestype=:scatter, size=(800,600), mc=:red, label=:none, ms=5, msc=:red)
title!("Stopping distance vs Speed")
xlabel!("Speed (mph)")
ylabel!("Stopping distance (ft)")
plot!(cars[!, "Speed"], β0 .+ β1*cars[!, "Speed"], label=:none, lc=:black, lw=3)
```

```julia
# Regressão linear
ols = lm(@formula(Dist ~ Speed), cars)
```
